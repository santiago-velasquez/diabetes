I want to deepen on this making a Montecarlo simulation to corroborate this hypothesis.

#### Montecarlo simulation to check relation between races and readmission

*Null hypothesis - h0:* I'll use as null hypothesis that every race would have similar numbers to the global readmission rate. It means that the race doesn't affect the probability of readmission. 

```python
np.mean(data['readmitted'])
```

So I'll simulate a lot of random experiments using the supposition (0.45) and check how is the behaviour of these. Expected: normal distribution around 0.45. Each experiment will have a size of 1000, that is the aproximate number of patients with minority races (Asians, hispanics, afroamericans, others)

```python
h0_prob = 0.45 # Our hypothesis of readmission
simulate = lambda p = h0_prob : (np.random.random(size=1000)<p).astype(int)
```

```python
simulated_exps = [np.mean(simulate()) for _ in range(100000)]
plt.hist(simulated_exps, bins=150);
plt.show()
```

Now I'll collect the observations of each race, and I'll check if these observations fits into the acceptance band of the experiment. I'll use the traditional band of acceptance of 2.5% -> 97.6%

```python
## Observations: readmission by race
asians = np.mean(data[data['race']=='Asian']['readmitted'])
hispanic = np.mean(data[data['race']=='Hispanic']['readmitted'])
caucasian = np.mean(data[data['race']=='Caucasian']['readmitted'])
afro_american = np.mean(data[data['race']=='AfricanAmerican']['readmitted'])
other = np.mean(data[data['race']=='Other']['readmitted'])
```

```python
plt.hist(simulated_exps, bins=150)
lw = 3
t0, t1 = np.percentile(simulated_exps, [2.5, 97.5])
plt.axvline(asians, color="red", label="Asian", lw=lw)
plt.axvline(hispanic, color="orange", label="Hispanic", lw=lw)
plt.axvline(caucasian, color="blue", label="Caucasian", lw=lw)
plt.axvline(afro_american, color="yellow", label="Afroamerican", lw=lw)
plt.axvline(other, color="violet", label="Other", lw=lw)



plt.axvline(t0, color="black", lw=2, ls="--", label="2.5, 97.6 percentiles")
plt.axvline(t1, color="black", lw=2, ls="--")
plt.legend() 
plt.grid()
plt.title("Montecarlo simulation")
plt.show()
```

So the asians and the other races are out of the acceptance band. That means that we can reject the null hypothesis for these races. So, for asians and other races there is a lower probability of being readmitted. For me it has cultural sense: asians are a very disciplined culture. They should have better control of their meds and the recommendations of the doctors. Also the eating habits of asians are well known for being very healthy, and the eating habits is a very important factor regarding to diabetes issues.

It's important to be aware of this relation of the races with readmission in order to keep this feature at the time of making predictive models, or manipulate the race column.

#### Creating new features that could be useful

I'll create a couple of new features based on the request, per patient, of hospital services. 1 -> On 1 year period.  2 -> On 10 years period. For the 1 year period I'll sum the features 'number_outpatient', 'number_emergency' and 'number_inpatient'. Those are features related to hospital services requesting on the last year.
for the 10 years period I'll check how many appeareances each patient has on the dataset. I expect that these new features have certain positive correlation with readmissions. 

```python
data['requests_1'] = data['number_outpatient'] + data['number_emergency'] + data['number_inpatient']
```

```python
sns.set_theme()

sns.distplot(data['requests_1'])
plt.show()
```

```python
requests_10 = data[data['patient_nbr'].duplicated()].patient_nbr.value_counts()
data['requests_10'] = data['patient_nbr'].apply(lambda x: requests_10[x] if x in requests_10 else 0)
```

```python
sns.set_theme()

sns.distplot(data['requests_10'])
plt.show()
```

```python
sns.histplot(data = data, x='requests_1', hue='readmitted', bins=30)
plt.show()
sns.histplot(data = data, x='requests_10', hue='readmitted', bins=30)
plt.show()
```

As we see on the images, there is present a direct relation between readmissions and requesting hospital services both in 1 year period as in 10 years period, so it's important to include these features that I've created at the time of making predictive models.
